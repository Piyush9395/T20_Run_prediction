# Create or update the 'current_score' column in 'df' to hold the cumulative sum of runs scored within each match, grouped by 'match_id'


Cumulative Sum (cumsum): This function in pandas adds up the values in a sequence progressively. For each element in the sequence, it computes the sum of all previous elements up to that point.

```python
df['current_score']=df.groupby('match_id')['runs'].cumsum()
```

# Convert 'player_dismissed' to binary (0 for no dismissal, 1 for dismissal), compute cumulative dismissals per match, and calculate the number of wickets left (10 minus cumulative dismissals)


```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x == '0' else 1)
df['player_dismissed']=df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()

df['wickets_left'] = 10 - df['player_dismissed']

#
```

# Calculate the current run rate by multiplying 'current_score' by 6 and dividing by 'balls_bowled', then store it in 'current_runrate'

# Convert 'runs' to numeric values (coerce errors to NaN), then compute the rolling sum of runs over a window of 30 deliveries for each match, and store the results in 'last_five'


```python
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')


groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id)['runs'].rolling(window=30).sum().values.tolist())
```

# Add a new column 'last_five' to 'df', containing the rolling sum of runs over the last 30 deliveries for each match


```python
df['last_five'] =  last_five
```

# Import required libraries for building and evaluating machine learning models


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score , mean_absolute_error
```

# Define a ColumnTransformer to apply OneHotEncoder to categorical columns 'batting_team', 'bowling_team', and 'city', while leaving other columns unchanged


```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city']),
], remainder='passthrough')  # Ensure passthrough to keep numerical columns
```

# Create a pipeline that first applies the ColumnTransformer for categorical encoding, then scales features with StandardScaler, and finally trains an XGBRegressor model with 1000 estimators, a learning rate of 0.2, and a max depth of 12


```python
pipe = Pipeline(steps=[
    ('step1',trf),          #handle categorical feature encoding and passthrough for other features.
    ('scaler',StandardScaler()),  #Standardizes the features to have mean 0 and variance 1.
    ('model',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

# Fit the pipeline to the training data, make predictions on the test set, and evaluate the model's performance using R2 score and mean absolute error
